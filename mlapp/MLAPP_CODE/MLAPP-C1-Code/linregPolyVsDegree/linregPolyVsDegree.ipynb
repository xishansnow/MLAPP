```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
from sklearn.linear_model import LinearRegression
plt.rcParams['figure.figsize'] = (15.0,5.0) # 设置figure_size尺寸
```

```python
%matplotlib inline
# 生成x数据
x=np.linspace(0,10,20)
y_net=np.power(x,2)+3  # 生成没有噪音的目标值y_net=x^2+3
```

```python
noise=4*np.random.randn(20)  # 生成服从正态分布的噪音数据，标准差为4，期望为0
```

```python
y=y_net+noise
print(y.shape,'\n',y[:5])  # 对y值进行噪音上的叠加，并查看部分数据
```

```python
# 对x进行形状上的改变，sklearn要求样本输入满足（n,d）的形式要求，n为样本数量，d为特征数量
X=x[:,np.newaxis]
X.shape
```

```python
# 使用1维特征对数据点进行拟合
lr=LinearRegression().fit(X,y)
```

```python
plt.plot(x,lr.predict(X))  # 绘制拟合的曲线
plt.plot(x,y,'o')          # 绘制样本点
```

```python
XX=np.tile(X,2)            # 对X的维度进行拓展，拓展后的X的数据形状为（n,2）
XX[:,1]=np.power(XX[:,1],2) # 第2列数据为x^2
lr2=LinearRegression().fit(XX,y) #使用拓展的特征进行拟合
lr2.coef_
```

```python
plt.plot(x,lr2.predict(XX)) # 绘制拟合后的曲线
plt.plot(x,y,'o')           # 绘制样本点
```

```python
ax1=plt.subplot(121)
ax2=plt.subplot(122)
ax1.plot(x,lr.predict(X))
ax1.plot(x,y,'o')
ax2.plot(x,lr2.predict(XX))
ax2.plot(x,y,'o')
```
