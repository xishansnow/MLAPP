```python
from scipy.stats import norm  # 导入正态分布
from scipy.stats import t     # 导入学生t分布
from scipy.stats import laplace  # 导入laplace分布
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(15,10)
```

```python
# 产生标准正态分布的数据
x1 = np.linspace(norm.ppf(0.001),norm.ppf(0.999),10000)
y1 = norm.pdf(x1)
logy1 = np.log(y1)
# 产生t分布的数据
df1,df2,df3 =1,3,5 # t 分布的自由度
x2 = np.linspace(t.ppf(0.001,df1),t.ppf(0.999,df1),10000)
y2 = t.pdf(x2,df1)
logy2 = np.log(y2)

x22 = np.linspace(t.ppf(0.001,df2),t.ppf(0.999,df2),10000)
y22 = t.pdf(x22,df2)
logy22 = np.log(y22)

x23 = np.linspace(t.ppf(0.001,df3),t.ppf(0.999,df3),10000)
y23 = t.pdf(x23,df3)
logy23 = np.log(y23)
# 产生laplace分布的数据
x3 = np.linspace(laplace.ppf(0.001,loc=0,scale=1/np.power(2,0.5)),
                laplace.ppf(0.999,loc=0,scale=1/np.power(2,0.5)),10000)
y3 = laplace.pdf(x3,loc=0,scale=1/np.power(2,0.5))
logy3 = np.log(y3)
print(x1,y1)
```

```python
#  进行绘图
ax1 = plt.subplot(221)
ax1.plot(x1,y1,color='black',linewidth=2,ls='-',label='Gauss')
ax1.plot(x2,y2,color='blue',linewidth=2,ls='-.',label='Student')
ax1.plot(x3,y3,color='red',linewidth=2,ls='--',label='Laplace')
ax1.set_xlim((-4,4))
ax1.set_ylim((0,0.8))
ax1.legend()
ax1.set_xlabel('(a)')

#  进行绘图
ax2 = plt.subplot(222)
ax2.plot(x1,logy1,color='black',linewidth=2,ls='-',label='Gauss')
ax2.plot(x2,logy2,color='blue',linewidth=2,ls='-.',label='Student')
ax2.plot(x3,logy3,color='red',linewidth=2,ls='--',label='Laplace')
ax2.set_xlim((-4,4))
ax2.set_ylim((-6,0))
ax2.legend()
ax2.set_xlabel('(b)')

ax3 = plt.subplot(223)
ax3.plot(x2,y2,color='blue',linewidth=2,ls='-.',label='Student v=1')
ax3.plot(x22,y22,color='green',linewidth=2,ls='-.',label='Student v=3')
ax3.plot(x23,y23,color='black',linewidth=2,ls='-.',label='Student v=5')
ax3.set_xlim((-8,8))
ax3.set_ylim((0,0.4))
ax3.legend()
ax3.set_xlabel('(c)')

ax4 = plt.subplot(224)
ax4.plot(x2,logy2,color='blue',linewidth=2,ls='-.',label='Student v=1')
ax4.plot(x22,logy22,color='green',linewidth=2,ls='-.',label='Student v=3')
ax4.plot(x23,logy23,color='black',linewidth=2,ls='-.',label='Student v=5')
ax4.set_xlim((-15,15))
ax4.set_ylim((-6,0))
ax4.legend()
ax4.set_xlabel('(d)')
```
