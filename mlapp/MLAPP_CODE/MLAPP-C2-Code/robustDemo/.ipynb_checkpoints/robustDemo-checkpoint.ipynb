```python
from scipy.stats import norm,t,laplace
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,5)
```

```python
# 产生服从正态分布的随机数
x=norm.rvs(loc=0,scale=1,size=100)
noise = np.array([8,9,10])
x_noise = np.append(x,noise)
bins=np.arange(-5,12)
ax1=plt.subplot(121)
ax2=plt.subplot(122)

result1=ax1.hist(x,bins,density=True,align='left',rwidth=0.7,facecolor='red')[0]
loc_norm,scale_norm=norm.fit(x)
df,loc_t,scale_t=t.fit(x)
loc_laplace,scale_laplace = laplace.fit(x)

result2=ax2.hist(x_noise,bins,density=True,align='left',rwidth=0.7,facecolor='red')[0]
loc_norm2,scale_norm2=norm.fit(x_noise)
df2,loc_t2,scale_t2=t.fit(x_noise)
loc_laplace2,scale_laplace2 = laplace.fit(x_noise)

# 绘制拟合后的曲线
x_norm = np.linspace(norm.ppf(0.001,loc=loc_norm,scale=scale_norm),
                    norm.ppf(0.999,loc=loc_norm,scale=scale_norm),1000)
ax1.plot(x_norm,norm.pdf(x_norm,loc=loc_norm,scale=scale_norm),color='black',linewidth=2,ls='-',label='Gaussian')

x_t = np.linspace(t.ppf(0.001,df=df,loc=loc_t,scale=scale_t),
                    t.ppf(0.999,df=df,loc=loc_t,scale=scale_t),1000)
ax1.plot(x_t,t.pdf(x_t,df=df,loc=loc_t,scale=scale_t),color='blue',linewidth=2,ls='-.',label='Student')

x_laplace = np.linspace(laplace.ppf(0.001,loc=loc_laplace,scale=scale_laplace),
                    norm.ppf(0.999,loc=loc_laplace,scale=scale_laplace),1000)
ax1.plot(x_laplace,laplace.pdf(x_laplace,loc=loc_laplace,scale=scale_laplace),color='green',linewidth=2,ls='--',label='laplace')

x_norm2 = np.linspace(norm.ppf(0.001,loc=loc_norm2,scale=scale_norm2),
                    norm.ppf(0.999,loc=loc_norm2,scale=scale_norm2),1000)
ax2.plot(x_norm2,norm.pdf(x_norm2,loc=loc_norm2,scale=scale_norm2),color='black',linewidth=2,ls='-',label='Gaussian')

x_t2 = np.linspace(t.ppf(0.001,df=df2,loc=loc_t2,scale=scale_t2),
                    t.ppf(0.999,df=df2,loc=loc_t2,scale=scale_t2),1000)
ax2.plot(x_t2,t.pdf(x_t2,df=df2,loc=loc_t2,scale=scale_t2),color='blue',linewidth=2,ls='-.',label='Student')

x_laplace2 = np.linspace(laplace.ppf(0.001,loc=loc_laplace2,scale=scale_laplace2),
                    norm.ppf(0.999,loc=loc_laplace2,scale=scale_laplace2),1000)
ax2.plot(x_laplace2,laplace.pdf(x_laplace2,loc=loc_laplace2,scale=scale_laplace2),color='green',linewidth=2,ls='--',label='laplace')

ax1.set_ylim((0,0.6))
ax2.set_ylim((0,0.6))
ax1.set_xlim((-6,11))
ax2.set_xlim((-6,11))
ax1.legend()
ax2.legend()
ax1.set_xlabel('(a)')
ax2.set_xlabel('(b)')
```
